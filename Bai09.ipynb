```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
src = 'D:/Image_TGMT/img06.jpg'
```

```python
img = Image.open(src)
plt.imshow(img)
```

```python
gray_img = img.convert('L')
plt.imshow(gray_img, cmap='gray')
```

```python
img_array = np.array(gray_img)
print(img_array)
```

```python
plt.figure()
plt.hist(img_array.flatten(), bins = 100)
plt.title('Biểu đồ Histogram')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng điểm ảnh')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(gray_img, origin="image")
plt.show()
```
