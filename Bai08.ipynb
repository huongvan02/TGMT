```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [20, 10, 30, 25, 45]
x1 = [1, 2, 3, 4, 5]
y1 = [15, 5, 40, 19, 30]
plt.plot(x, y, 'go--', label='Dữ liệu 1')
plt.plot(x1, y1, marker='o', color='b', linestyle=':', label='Dữ liệu 2')
plt.legend()
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 5, 25, 14, 18]
plt.bar(categories, values, color='g', alpha=0.6)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [35, 15, 20, 30]
my_colors = ['red', 'green', 'blue', 'yellow']
plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [20, 10, 30, 24, 45]
plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Biểu đồ phân tán')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend()
plt.show()
```

```python
data = [10, 20, 30, 33, 35, 45, 45, 51, 55]
plt.boxplot(data)
plt.title('Biểu đồ hộp')
plt.ylabel('Giá trị')
plt.show()
```

```python
data = [15, 20, 22, 30, 35, 40, 45]
plt.violinplot(data)
plt.title('Biểu đồ violin')
plt.ylabel('Giá trị')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text = "Python is an amazing language for data visualization and word clouds."
wordcloud = WordCloud(width=800, height=400).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Biểu đồ đám mây từ văn bản')
plt.show()
```

```python
import numpy as np
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 6, 4]
N = len(categories)
values += values[:1]
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]
plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```
